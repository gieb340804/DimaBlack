### Метод сортировки Сортировка перестановками (permutation sort)

```python
def sort_even_odd(arr):
  """
  Сортирует целочисленный массив так, что все четные числа идут перед нечетными.
  """

  # Разделяем массив на четные и нечетные элементы.
  even = []
  odd = []
  for num in arr:
    if num % 2 == 0:
      even.append(num)
    else:
      odd.append(num)

  # Сортируем четный массив.
  even.sort()

  # Сортируем нечетный массив.
  odd.sort()

  # Объединяем четный и нечетный массивы.
  return even + odd
    
print(sort_even_odd([34, 51, 73, 61, 18, 42, 65, 38, 90, 43, 20, 62, 10, 64, 38, 75, 53, 66, 80, 69, 46, 87, 81, 48, 94, 63, 100, 52, 35, 54, 52, 51, 84, 92, 94, 2, 81, 14, 26, 50, 43, 25, 54, 2, 90, 94, 72, 71, 31, 5, 46, 9, 28, 45, 16, 99, 31, 50, 47, 57, 44, 72, 9, 5, 81, 10, 49, 62, 17, 29, 57, 8, 50, 35, 42, 93, 58, 100, 44, 87, 38, 40, 89, 25, 25, 59, 82, 11, 61, 23, 70, 9, 37, 52, 12, 80, 43, 59, 8, 70]))
```
